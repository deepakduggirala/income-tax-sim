```python
%matplotlib notebook

import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
sns.set(style="darkgrid")
```

```python
income =np.arange(1, 5010000, 10000)

def taxf(x):
    if x < 250000:
        return 0
    if x < 500000:
        return 0.05*(x-250000)
    if x < 1000000:
        return 0.05*(250000) + 0.2*(x-500000)
    return 0.05*(250000) + 0.2*(500000) + 0.3*(x-1000000)

tax = [taxf(i) for i in income]
net_income = income - tax

percent_tax = tax/income*100

fig, (ax1,ax2) = plt.subplots(2,1, figsize=(8,9))
ax1.plot(income, tax, income, net_income, income, income)
ax2.plot(income, percent_tax)
ax2.set_ylabel('Percent Tax')
ax2.set_xlabel('Income')
```
